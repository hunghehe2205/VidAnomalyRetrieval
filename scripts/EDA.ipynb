```python
import json
from collections import defaultdict
```

```python
# Paths
train_path = "/Users/hunghehe2205/Projects/VidAnomalyRetrieval/data/T2V_VAR/ucf_crime_train.json"
test_path = "/Users/hunghehe2205/Projects/VidAnomalyRetrieval/data/T2V_VAR/ucf_crime_test.json"

# Load data
def load_json(path):
    with open(path, "r", encoding="utf-8") as f:
        return json.load(f)

train_data = load_json(train_path)
test_data = load_json(test_path)
```

```python
# Combine
all_data = train_data + test_data

print(f"Total samples: {len(all_data)}")

# Extract queries
queries = [item["English Text"].strip() for item in all_data if "English Text" in item]

# Count unique
unique_queries = set(queries)
print(f"Unique queries: {len(unique_queries)}")

# Find duplicates
query_to_videos = defaultdict(list)

for item in all_data:
    query = item["English Text"].strip()
    video = item["Video Name"]
    query_to_videos[query].append(video)

# Filter duplicates
duplicates = {q: vids for q, vids in query_to_videos.items() if len(vids) > 1}

print(f"Number of duplicated queries: {len(duplicates)}")

# Show results
for i, (query, videos) in enumerate(duplicates.items()):
    print(f"\nDuplicate #{i+1}")
    print(f"Query: {query}")
    print("Videos:")
    for v in videos:
        print(f"  - {v}")
```
